```python
import numpy as np
import pandas as pd
from sklearn.datasets import load_boston
```

```python
data = load_boston()
```

```python
dir(data)
```

```python
data = pd.DataFrame(data.data, columns = data.feature_names)

data.head()
```

```python
corr_matrix = data.corr()
```

```python
corr_matrix
```

```python
upp_tri = corr_matrix.where(np.triu(corr_matrix, k = 1).astype(np.bool))
```

```python
corr_cols = [col for col in upp_tri.columns if any(abs(upp_tri[col]) > 0.7)]

corr_cols
```
