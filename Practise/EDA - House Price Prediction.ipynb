## Missing values

```python
# 1. step make the list of features which has missing values
features_with_na = [features for features in dframe.columns if dframe[features].isnull().sum() > 1]

for feature in features_with_na:
    print(feature, "\t\t", np.round(dframe[feature].isnull().mean(),4), "\t \t % MISSING VALUES")
```

```python
## Since there are many missing values, We have to find the relationship between the features and Targer variable
```

```python
for feature in features_with_na:
    data = dframe.copy()
    data[feature] = np.where(data[feature].isnull(),1,0)
    data.groupby(feature)['SalePrice'].median().plot.bar()
    plt.title(feature)
    plt.show()
```

 Here With the relation between the missing values and the dependent variable is clearly visible.So We need to replace these nan values with something meaningful which we will do in the Feature Engineering section
 
 From the data set, The feature ID is not required and that can be dropped.

### Exploratory Data Analysis Part : 2 

If the data is skewed, We have to covert the data to the logarithmic transformation

```python
for feature in continueous_features:
    data = dframe.copy()
    if 0 in data[feature].unique():
        pass
    else:
        data[feature] = np.log(data[feature])
        data['SalePrice'] = np.log(data['SalePrice'])
        plt.scatter(data[feature], data['SalePrice'])
        plt.xlabel(feature)
        plt.ylabel("Sale Price")
        plt.show()
```

## Outliers

```python
for feature in continueous_features:
    data = dframe.copy()
    if 0 in data[feature].unique():
        pass
    else:
        data[feature] = np.log(data[feature])
        data.boxplot(column = feature)
        plt.ylabel("Sales Price")
        plt.show()
```
