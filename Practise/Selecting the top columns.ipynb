```python
import numpy as np
import pandas as pd
from sklearn.feature_selection import mutual_info_regression, SelectPercentile
```

```python
data = pd.read_csv("D:\Data Science\Datasets\housetrain.csv")

data.head()
```

```python
X = data.drop('SalePrice', axis = 1)
y = data['SalePrice']
```

```python
X.head()
```

```python
X.info()
```

```python
X.isnull().sum()
```

```python
X = X.fillna(0)
```

```python
X.isnull().sum()
```

```python
numeric_lst=['int16', 'int32', 'int64', 'float16', 'float32', 'float64']
numerical_cols = list(X.select_dtypes(include=numeric_lst).columns)
```

```python
X_numeric_cols = X[numerical_cols]
```

```python
mutual_info = mutual_info_regression(X_numeric_cols, y)

mutual_info
```

```python
mutual_info = pd.Series(mutual_info, index = X_numeric_cols.columns)

mutual_info
```

```python
selected_top_columns = SelectPercentile(mutual_info_regression, percentile = 20)

selected_top_columns.fit(X_numeric_cols, y)
```

```python
selected_top_columns.get_support()
```
